```python
import os
import subprocess

def run_command(command):
    """Run a shell command and return the output."""
    try:
        result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)
        return result.stdout.strip()
    except subprocess.CalledProcessError as e:
        print(f"Error running command: {command}\n{e.stderr}")
        return None

def check_ollama_installed():
    """Check if Ollama is installed."""
    print("üîç Checking if Ollama is installed...")
    result = run_command("which ollama")
    return result is not None

def install_ollama():
    """Install Ollama on macOS."""
    print("üì• Installing Ollama...")
    run_command("curl -fsSL https://ollama.com/install.sh | sh")

def start_ollama_server():
    """Start the Ollama server in the background."""
    print("üöÄ Starting Ollama server...")
    run_command("ollama serve &")

def check_installed_models():
    """Check which models are installed."""
    print("üîç Checking installed models...")
    models = run_command("ollama list")
    return models if models else ""

def download_model(model_name="mistral"):
    """Download a model for Ollama."""
    print(f"üì• Downloading model: {model_name}...")
    run_command(f"ollama pull {model_name}")

def verify_ollama():
    """Verify that Ollama is running and models are available."""
    print("‚úÖ Verifying Ollama setup...")
    ollama_version = run_command("ollama --version")
    models = check_installed_models()
    print(f"Ollama Version: {ollama_version}")
    print(f"Installed Models: {models}")

if __name__ == "__main__":
    # Step 1: Check if Ollama is installed
    if not check_ollama_installed():
        install_ollama()
    
    # Step 2: Start Ollama server
    start_ollama_server()

    # Step 3: Check for installed models
    installed_models = check_installed_models()
    if "mistral" not in installed_models.splitlines():
        download_model("mistral")  # You can change this to "llama2" or "gemma"

    # Step 4: Verify everything
    verify_ollama()

    print("üéâ Ollama setup is complete! You can now use it in Python.")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

# Choose DeepSeek model (7B version)
MODEL_NAME = "deepseek-ai/deepseek-llm-7b-chat"

print(f"üì• Loading model: {MODEL_NAME}... (This may take a few minutes)")

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,  # Use FP16 for efficiency
    device_map="auto"  # Automatically uses GPU if available
)

# Create a text-generation pipeline
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Function to chat with DeepSeek
def chat_with_deepseek(prompt, max_length=200):
    response = llm_pipeline(prompt, max_new_tokens=max_length, do_sample=True, temperature=0.7)
    return response[0]['generated_text']

# Chatbot loop
print("ü§ñ DeepSeek Chatbot is ready! Type 'exit' to quit.")
while True:
    query = input("You: ")
    if query.lower() == "exit":
        print("üëã Goodbye!")
        break
    response = chat_with_deepseek(query)
    print("ü§ñ DeepSeek:", response)
```

```python
import ollama
import streamlit as st
import time


def stream_Data(text,delay:float=0.02):
    for word in text.split():
        yield word + ' '
        time.sleep(delay)

# input
prompt = st.chat_input("Enter your text here")

if prompt:
    with st.chat_message("user"):
        st.write(prompt)

    # processing
    with st.spinner("Processing..."):
        result = ollama.chat(model = "tinyllama",messages=[{
            "role": "user",
            "content": prompt
        }])
        response = result["message"]["content"]
        #st.write_stream(response)
        st.write(stream_Data(response))
```
